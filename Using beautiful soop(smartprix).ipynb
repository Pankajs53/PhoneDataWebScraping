```python
with open('smartprixphones.html','r',encoding='utf-8') as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
soup=BeautifulSoup(html,'lxml')
```

```python
soup.prettify()
```

```python
//*[@id="app"]/main/div[1]/div[2]/div[2]/div[951]
```

```python
# total no of phones
containers=soup.find_all('div',{'class':'sm-product has-tag oos has-features has-actions'})
```

```python
name=[]
price=[]
specs=[]

for i in containers:
    name.append(i.find('h2').text)
    price.append(i.find('span',{'class':'price'}).text)
    try:
        specs.append(i.find('div',{'class':'score rank-2-bg'}).find('b').text)
    except:
        specs.append(np.nan)
```

```python
df=pd.DataFrame({
    'model':name,
    'price':price,
    'specs':specs
})

df
```
